# DATA CLEANING FOR MS EXCEL :

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head(5)
```

```python
print("\nMissing values per column:\n")
print(df.isnull().sum())

df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)

df.drop_duplicates(inplace=True)

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract Year and Month from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()

# Extract numeric duration (for Movies only)
def extract_duration(x):
    try:
        if 'min' in x:
            return int(x.split()[0])
        else:
            return np.nan
    except:
        return np.nan

df['duration_num'] = df['duration'].apply(extract_duration)
```

```python
df['primary_country'] = df['country'].apply(lambda x: x.split(',')[0] if x != 'Unknown' else 'Unknown')
```

```python
df['primary_listed_in'] = df['listed_in'].apply(lambda x: x.split(',')[0] if x != 'Unknown' else 'Unknown')
```

```python
df['primary_director'] = df['director'].apply(lambda x: x.split(',')[0] if x != 'Unknown' else 'Unknown')
```

```python
df['primary_cast'] = df['cast'].apply(lambda x: x.split(',')[0] if x != 'Unknown' else 'Unknown')
```

```python
df['primary_rating'] = df['rating'].apply(lambda x: x.split(',')[0] if x != 'Unknown' else 'Unknown')
```
